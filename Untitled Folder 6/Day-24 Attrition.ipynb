# Project 2  Attrittion Rate Analysis

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
```

```python
attrition=pd.read_csv("general_data (1).csv")
attrition.head()
```

```python
label=preprocessing.LabelEncoder()
attrition["Attrition"]=label.fit_transform(attrition["Attrition"])
attrition["Gender"]=label.fit_transform(attrition["Gender"])
attrition.head()
```

```python
attrition.isnull().sum()
```

```python
print(attrition["NumCompaniesWorked"].mode())
print(attrition["TotalWorkingYears"].mean())
```

```python
attrition["NumCompaniesWorked"]=np.where(attrition["NumCompaniesWorked"].isnull(),1.0,attrition["NumCompaniesWorked"])
attrition["TotalWorkingYears"]=np.where(attrition["TotalWorkingYears"].isnull(),11,attrition["TotalWorkingYears"])
```

# Random Forest Test

```python
 attrition.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
faetures=['Age','DistanceFromHome','Education','Gender','NumCompaniesWorked','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsWithCurrManager','YearsSinceLastPromotion','PercentSalaryHike']
```

```python
rf_model.fit(X=attrition[faetures],y=attrition["Attrition"])
print("OOB Accuracy:")
print(rf_model.oob_score_)
```

```python
for faetures,imp in zip(faetures,rf_model.feature_importances_):
    print(faetures,imp)
```

# Descision Tree

```python
predictors=pd.DataFrame([attrition["Age"],attrition["DistanceFromHome"],attrition["TotalWorkingYears"],attrition["PercentSalaryHike"]]).T
tree_model=tree.DecisionTreeClassifier(max_depth=6)
tree_model.fit(X=predictors,y=attrition['Attrition'])
```

```python
with open("AttriTree.dot",'w') as f:
    f=tree.export_graphviz(tree_model,feature_names=['Age','DistanceFromHome','TotalWorkingYears','PercentSalaryHike'],out_file=f)
```

```python
from sklearn.tree import export_text
r=export_text(tree_model,feature_names=['Age','DistanceFromHome','TotalWorkingYears','PercentSalaryHike'])
print(r)
```
