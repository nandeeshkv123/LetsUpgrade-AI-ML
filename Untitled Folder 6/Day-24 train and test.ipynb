```python
import pandas as pd                   # for loading data frame
```

```python
import numpy as np                       #for storing the data in  array and matrix form
```

```python
from sklearn import tree                   # sklearn used for machine larning algoritm and imported tree as function
```

```python
from sklearn import preprocessing      # preprocesing used to convert text into numerical and any missing value.
```

```python
titanic_train=pd.read_csv("train.csv")
titanic_train.head()
```

```python
titanic_train['Age'].mean()
```

```python
new_age_var=np.where(titanic_train['Age'].isnull(),32,titanic_train['Age'])
```

```python
titanic_train['Age']=new_age_var
```

```python
label_encoder=preprocessing.LabelEncoder()               #labelencoder is used to convert text into 0 and 1.
```

```python
titanic_train["Sex"]=label_encoder.fit_transform(titanic_train["Sex"])
```

```python
tree_model=tree.DecisionTreeClassifier()
```

```python
predictors=pd.DataFrame([titanic_train["Sex"],titanic_train["Age"],titanic_train["Fare"]]).T
```

```python
tree_model.fit(X=predictors,y=titanic_train["Survived"])
```

```python
with open("Dtree1.dot",'w') as f:
    f= tree.export_graphviz(tree_model,feature_names=['Sex','Age','Fare'],out_file=f)
```

```python
from sklearn.tree import export_text
r=export_text(tree_model,feature_names=['Sex','Age','Fare'])
print(r)
```
